# Q2. Check for missing values in all columns and replace them with the appropriate metric
# (Mean/Median/Mode)

```python
# check missing values
df.isnull().sum() 
```

```python
df['Description'].fillna(df['Description'].mode()[0], inplace=True)    # 'Description' null fill with mode

df['CustomerID'].fillna(df['CustomerID'].mode()[0], inplace=True)      # 'CustomerID' null fill with mode 
```

```python
# Their now No missing values
df.isnull().sum()              # All o means no missing 
```

# Q3. Remove duplicate rows

```python
# Duplicates removed
df.drop_duplicates(inplace=True)
```

```python
# No duplicates now
df.duplicated().sum()    
```

# Q4. Remove rows which have negative values in Quantity column

```python
df = df[df['Quantity']>0]
```

# Q5. Add the columns - Month, Day and Hour for the invoice

```python
# convert coln into date format
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate']) 
df['InvoiceDate'].dtype
```

```python
# create day, mnth, hr
df['day']=df['InvoiceDate'].dt.day

df['Month']=df['InvoiceDate'].dt.month

df['Hour']=df['InvoiceDate'].dt.hour
```

# Q8. How much money spent by the customers?

```python
df.head()
```

```python
# calculate amonut spent column for each entry 
df['amount_spent'] = df['Quantity']*df['UnitPrice']  
df.head()
```

# Q10. How many orders per month?

```python
ax = df.groupby('InvoiceNo')['Month'].unique().value_counts().sort_index().plot(kind='bar',figsize=(6,6))
ax.set_xlabel('Month',fontsize=15)
ax.set_ylabel('Number of Orders',fontsize=15)
ax.set_title('Number of orders for different Months',fontsize=15)

plt.show()
```

# Q11. How many orders per day?

```python
ax = df.groupby('InvoiceNo')['day'].unique().value_counts().sort_index().plot(kind='bar',figsize=(15,6))
ax.set_xlabel('Month',fontsize=15)
ax.set_ylabel('Number of Orders',fontsize=15)
ax.set_title('Number of orders per day',fontsize=15)

plt.show()
```

# Q12. How many orders per hour?

```python
df.groupby('InvoiceNo')['Hour'].unique().value_counts().iloc[:-1].sort_index()
```

```python
ax = df.groupby('InvoiceNo')['Hour'].unique().value_counts().iloc[:-1].sort_index().plot(kind='bar',figsize=(15,6))
ax.set_xlabel('Hour',fontsize=15)
ax.set_ylabel('No of orders',fontsize=15)
ax.set_title('No of orders for different hours',fontsize=15)
ax.set_xticklabels(range(6,21),rotation=360, fontsize=15)
plt.show()
```

# Q13. How many orders for each country?

```python
country_orders = df.groupby('Country')['InvoiceNo'].count().sort_values() 
plt.subplots(figsize=(15,8))
country_orders.plot(kind='barh',fontsize=12,figsize=(15,8))
plt.xlabel('Number of orders',fontsize=15)
plt.ylabel('Country',fontsize=15)
plt.title('No. of orders for countries',fontsize=15)
plt.show()
```

```python
# delete uk  as it is already high order no use, to see other country order details clearly.
del country_orders['United Kingdom']
plt.subplots(figsize=(15,8))
country_orders.plot(kind='barh',fontsize=12,figsize=(15,8))
plt.xlabel('Number of orders',fontsize=15)
plt.ylabel('Country',fontsize=15)
plt.title('No. of orders for countries',fontsize=15)
plt.show()
```
